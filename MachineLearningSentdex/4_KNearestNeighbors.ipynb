# K Nearest Neighbors

```python
"""
K value is usually odd number.
KNN is not an efficient algorithm. KNN isn't actually train and test the data.
For small datasets KNN is good but the larger the dataset,
the worse this algorithm runs.
"""
```

```python
import numpy as np 
import pandas as pd
from sklearn import preprocessing, model_selection, neighbors
```

```python
df = pd.read_csv("Data/breast-cancer-wisconsin.data")
"""
Editng single missing data
In .names file we saw the missing data is shown with ?
"""
df.replace('?', -99999, inplace=True)
"""id column is pretty useless. Also it ruins the KNN accuracy."""
df.drop(columns='id', inplace=True)
df.head()
```

```python
X = np.array(df.drop(columns=["class"]))
y = np.array(df["class"])

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)
```

```python
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)
accuracy
```

```python
"""Predicting a data"""
example_measures = np.array([[4, 2, 1, 1, 1, 2, 3, 2, 1], [4, 2, 1, 2, 2, 2, 3, 2, 1]])
example_measures = example_measures.reshape(len(example_measures), -1)

prediction = clf.predict(example_measures)
print(prediction)
```

### Calculating Overall Accuracy

```python
accuracies = list()

for i in range(25):
    df = pd.read_csv("Data/breast-cancer-wisconsin.data")
    """
    Editng single missing data
    In .names file we saw the missing data is shown with ?
    """
    df.replace('?', -99999, inplace=True)
    """id column is pretty useless. Also it ruins the KNN accuracy."""
    df.drop(columns='id', inplace=True)

    X = np.array(df.drop(columns=["class"]))
    y = np.array(df["class"])

    X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)
    """n_jobs: The number of parallel jobs to run for neighbors search."""
    clf = neighbors.KNeighborsClassifier(n_jobs=-1)
    clf.fit(X_train, y_train)

    accuracy = clf.score(X_test, y_test)
    
    accuracies.append(accuracy)

print("Total Accuracy:", sum(accuracies)/len(accuracies))
```
